##### MINI TASK: FILTERING MODULE

```python
def filter_by_points(df, min_points):
    return df[df["points"] >= min_points]

def filter_by_price(df, max_price):
    return df[df["price"] <= max_price]

def filter_by_country(df, countries):
    return df[df["country"].isin(countries)]

def search_title(df, keyword):
    return df[df["title"].str.contains(keyword, case=False, na=False)]
```

#### Combine Filters

```python
top_wines = filter_by_points(df, 90)
top_wines = filter_by_price(top_wines, 40)
top_wines = filter_by_country(top_wines, ["US", "France"])

top_wines.head()
```
