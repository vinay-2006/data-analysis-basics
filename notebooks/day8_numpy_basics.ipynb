```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Version checking 

```python
print("NumPy:", np.__version__)
print("Matplotlib:", plt.matplotlib.__version__)
```

#### Array Creation

```python
a = np.array([1, 2, 3])
b = np.zeros((3, 3))
c = np.ones((2, 4))
d = np.arange(10)
e = np.linspace(0, 1, 5)

a, b, c, d, e
```

#### Reshaping 

```python
x = np.arange(12)
x_2d = x.reshape(3, 4)
x_3d = x.reshape(2, 2, 3)

x, x_2d, x_3d
```

#### Indexing and Slicing 

```python
matrix = np.arange(25).reshape(5, 5)

row_2 = matrix[2]
col_3 = matrix[:, 3]
sub = matrix[1:4, 1:4]

matrix, row_2, col_3, sub
```

####  Random Arrays 

```python
rand_matrix = np.random.rand(3, 3)
rand_ints = np.random.randint(1, 10, size=(4, 4))

rand_matrix, rand_ints
```

#### Generating Heatmap

```python
heat = np.random.randint(1, 100, (10, 10))
heat
```

#### Basics Heatmap

```python
plt.imshow(heat, cmap="viridis")
plt.colorbar()
plt.title("10Ã—10 Heatmap")
plt.show()
```

#### Enhanced Heatmap

```python
plt.figure(figsize=(6, 5))
plt.imshow(heat, cmap="inferno")
plt.colorbar(label="Intensity")
plt.xticks(range(10))
plt.yticks(range(10))
plt.title("Enhanced Heatmap")
plt.tight_layout()
plt.show()
```

#### Save Heatmap

```python
plt.figure(figsize=(6, 5))
plt.imshow(heat, cmap="inferno")
plt.colorbar(label="Intensity")
plt.xticks(range(10))
plt.yticks(range(10))
plt.title("Enhanced Heatmap")
plt.tight_layout()
plt.savefig("heatmap.png", dpi=300)
plt.show()
```
